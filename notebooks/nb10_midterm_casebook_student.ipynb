## Before the Midterm: Preparation Checklist

The five cases above are the *fields* the midterm tests on. The *toolkit* it tests with is everything you've built across NB01 through NB09. Walk through this checklist before exam time.

**Toolkit you should have at fingertip recall:**

- **Problem framing**: regression vs. classification; what makes a target a *target* vs. an *outcome that arrives later than the decision*
- **Metric selection by case**: MAE / RMSE / R² for regression; accuracy / precision / recall / F1 / ROC-AUC / PR-AUC for classification — including which metric is the trap at heavy imbalance and why
- **Split strategies**: 60/20/20 with `random_state=474` — stratified for classification (preserves the positive-class rate per fold) and random shuffle for regression (rows are independent observations); plain `KFold` for regression CV, `StratifiedKFold` for classification CV
- **Baselines**: `DummyRegressor`/`DummyClassifier` (mean / most_frequent), segment-aware mean (per category for regression, per cohort for classification)
- **Pipelines**: `ColumnTransformer` with numerics → `StandardScaler` and categoricals → `OneHotEncoder(handle_unknown='ignore')`; the rule that **any fit step belongs inside the pipeline**
- **Leakage flavors**: post-decision features, future-window features, target encoding fit on full data
- **Cross-validation with 95% CI**: 5-fold CV mean ± Student's *t* CI; INSIDE / ABOVE / BELOW verdict against single-split numbers
- **CI-overlap rule**: when two candidates' confidence intervals overlap, prefer the simpler / more interpretable / faster model
- **Threshold tuning under cost asymmetry**: sweep cost across thresholds and pick the minimum; threshold below 0.5 when missing a positive is far more expensive than a false alarm; above 0.5 when the labeled positive is the *desired* outcome (e.g., AstraOrbit)
- **Imbalance discipline**: at heavy imbalance the accuracy paradox is real — a "predict majority" baseline already scores 88% / 97.5% / 99.82%; PR-AUC and recall on the minority class are the headline metrics

## During the Midterm: Test-Taking Tips

**Read the briefing first, in full.** Every detail in the briefing — the dataset size, the class balance, the cost weights, the split convention — is potentially load-bearing for the question. The briefings average \~600 words. A 3–5 minute read pays for itself across the question's six or seven possible reasoning paths.

**Anchor on the manager role.** The question stem will explicitly say *"As VP of Fab Operations, …"* or *"As Chief Risk Officer, …"*. Pause for a beat and put yourself there. The right option is not the one that maximizes a CV metric — it is the one a *thoughtful manager* in that role would defend in a review meeting.

**Spot the failure archetype in each distractor.** Three of the four options are wrong on purpose. They are wrong in *characteristic* ways — rubber-stamp approval ("0.97 is great, ship it"), paranoid rejection ("kill the project"), approval for a wrong reason ("approve because the new feature is novel"), over-engineering ("add 200 more features"), misplaced trust ("the senior data scientist said so"), convention confusion ("StratifiedKFold is for time series"), or premature optimization ("tune hyperparameters before establishing a baseline"). When you can name the archetype in three options, the remaining option is your answer.

**Watch for post-decision features.** Any feature whose value is recorded *after* the moment the model is asked to predict — `next_hour_temperature` (a forecast at decision time, not the realized value), `final_test_failure_rate` (measured after the wafer is processed), `chargeback_dispute_filed` (filed days after authorization), `weather_at_landing_time` (the actual landing-burn weather), `support_ticket_volume_first_week` (truncated for users who convert early) — is target leakage. The CV score will look great. Reject.

**Trust CI overlap.** If two candidate models have CIs that overlap, the apparent mean difference is not statistically distinguishable. Pick the simpler, more interpretable, faster, or more auditable model — even if its mean is slightly lower.

**Pace yourself.** If you're 18 minutes into a case, mark your best guess and move on — you cannot return, and the next case may have an obvious answer that recovers the time.

**Don't second-guess after submission.** Once you click submit, the page is locked. Move forward, fully present for the next case. Re-running the prior decision in your head only costs you minutes on the case in front of you.
