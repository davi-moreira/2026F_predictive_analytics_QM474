Your M2 baseline is the floor your M3 complex model has to clear. Replicate it cleanly so the comparison in §1b has something fair to land against:

## 5. M3 Component §1b — More Complex Model + Tuning + Selection  *(35 pts combined)*

§1b is the heart of M3 — it carries **35 of the 100 points** by itself. Four sub-deliverables to land.

### 5.1 Model choice and justification  *(8 pts)*

Pick a more flexible model family that fits the prediction goal. Standard options for this course are **Random Forest**, **Gradient Boosting**, **SVM**, or something equally appropriate for your data. The justification is one paragraph — not one sentence. Connect the family's strengths (non-linearities, interactions, locally piecewise behavior) to what you actually expect your dataset to look like. A vague *"random forests are often good"* won't earn the full eight points; *"we expect non-monotonic effects between square footage and price, and a random forest can carve those out without us hand-coding the breakpoints"* will.

### 5.2 Hyperparameter tuning with cross-validation  *(12 pts)*

Set up a grid (or a randomized-search distribution) over the hyperparameter values that actually matter for your model family — tree depth and number of estimators for forests, learning rate and number of estimators for boosting, regularization strength and kernel for SVMs. Run it through **`GridSearchCV` or `RandomizedSearchCV` with 5- or 10-fold CV** on the training fold only. Report the results as either a table or a plot of the CV metric across configurations. The grader has to be able to *see* the search you did.

### 5.3 Model selection + final-training step  *(10 pts)*

Take the best hyperparameter combination from §5.2 and compare its CV mean + 95% CI directly against the M2 baseline's. Apply the **CI-overlap rule** from nb08:

- If the two CIs are **disjoint**, your complex model's gain is statistically real — adopt it as your champion.
- If the two CIs **overlap**, prefer the simpler baseline. Honor the rule unless there is a clear operational argument to keep the complex model (e.g., the baseline genuinely cannot handle a feature type the complex model can).

Once you have a champion, **retrain the chosen Pipeline on the full training fold (train + validation rows together)** so the model that ends up in the report has learned from every row of development data available to it.

### 5.4 Comparison narrative vs. the baseline  *(5 pts)*

In one paragraph, explain whether the complexity was worth it. Show both CIs side by side. Be honest — if the CIs overlap and your baseline wins, *say so*. *"More complexity didn't help here, and we have evidence of that"* is a perfectly defensible M3 conclusion and shows you understand the rule. Pretending a tiny mean improvement matters when the CIs overlap is the kind of thing the rubric flags.

## 6. M3 Component §1c — Required Visualizations  *(20 pts)*

Every figure below goes in the report with axis labels, units where they apply, a clear legend, and a 1–2 sentence caption explaining what to read off the figure. The grader's eye goes to figures first — a captioned, well-labeled chart that your prose actually references is worth a lot more than a bare PNG in the appendix.

**Required for every project:**

- **Hyperparameter-search plot.** CV metric versus the hyperparameter value(s). Use a line plot for a 1-D grid, a heatmap for a 2-D grid, or a boxplot of per-fold scores across configurations. Mark the chosen best point visually (a star, a circled cell, whatever reads cleanly).
- **Model-comparison bar chart.** Two bars side by side — M2 baseline vs M3 complex champion — each carrying its **CV mean with 95% Student's *t* CI error bars**. This is the figure your CI-overlap-rule decision rests on; the rubric will look here first.
- **Feature importance / coefficient plot.** A horizontal bar chart of the top features for your champion. Use signed coefficients for linear / logistic models; permutation importance or impurity importance for tree-based models.

**For regression projects:**

- **Predicted-vs-actual scatter plot** with a 45° reference line (`y = x`).
- **Residual plot** (residuals vs predicted) with a horizontal `y = 0` reference line.
- **Optional but recommended:** **RMSE-by-quintile bar chart** in stakeholder units (e.g., USD). If you can quote *"the model is typically off by USD X on a USD Y home"* in your report, the deployment council can read it without translation.

**For classification projects:**

- **Confusion matrix** at your chosen operating threshold (`ConfusionMatrixDisplay`).
- **ROC curve with AUC annotation** plus **Precision-Recall curve with PR-AUC annotation**.
- **Optional but recommended:** **Reliability diagram + Brier score** if your model's probabilities feed a downstream decision threshold.

Each figure earns its points on labeling, captioning, and integration into the narrative — not just on rendering. A figure that your prose refers to by name is worth four times an uncaptioned PNG.

## 7. M3 Component §2 — Draft Abstract  *(\~250 words, 15 pts)*

The abstract is the lead paragraph of your M4 poster. Treat it like a press release for your project — every sentence pulls weight.

**The six elements your abstract must contain, in this order:**

1. **Project title.** Concise and informative. *If your project uses a synthetic generated dataset, the title must say so explicitly.*
2. **Prediction problem, framed as a question with a "?".** Example: *"Can six-month customer-churn risk be predicted from transaction history and engagement metrics?"*
3. **Prediction goal and motivation.** One or two sentences: what you're predicting and why someone should care.
4. **Methodology and tools.** A short summary of your data preparation, your analytical methods (baseline + complex model + tuning + CV protocol), and the tools you used (sklearn, etc.).
5. **Key findings / expected contributions.** A brief overview of preliminary findings or anticipated contributions.
6. **Broader implications.** One sentence on how the project informs business practice or contributes to the broader field of predictive analytics.

**Two pieces of advice that save groups real time:**

- **Write the abstract LAST.** Draft the modeling section first; the abstract synthesizes the numbers *after* they settle. Writing it before the CV CI is known produces vague, hedge-filled prose that the rubric penalizes.
- **Read it aloud, three times.** If you stumble on a sentence, your audience will too. Three rounds of revision is the floor, not the ceiling.

## 9. Tips and Common Pitfalls

- **Tune on the training fold only.** `GridSearchCV` runs CV inside the training fold; the held-out test set stays in the lockbox until M4. Reporting *"best test MAE at alpha=10"* after evaluating all alphas on the test set is a textbook leakage failure — graders look for this.
- **Show the search.** A hyperparameter table or a plot of CV metric vs hyperparameter value is the easiest way to convince the grader that your search was systematic and not cherry-picked.
- **Honor the CI-overlap rule.** If your complex model's CI overlaps the baseline's, the simpler model wins. That's a finding, not a failure — *"more complexity didn't help here, and we have evidence of that"* is a clean M3 conclusion.
- **Write the abstract last.** The numbers settle first, then the prose. Trying to write the abstract before the modeling is done produces vague language that the rubric penalizes under §2.
- **The abstract IS the poster lead.** Polish it. Read it aloud. Three rounds of revision is the floor, not the ceiling.
- **Required visualizations are graded as a set.** The hyperparameter-search plot, the model-comparison bar chart with 95% CI error bars, the feature-importance plot, and the regression / classification diagnostic figures are graded under §1c (20 points). Drafting them while the modeling section is still open catches issues early.
- **Don't chase the public Kaggle leaderboard.** The private leaderboard scores the held-out remainder of the test set. Models that overfit the public split routinely drop 10+ ranks when the private leaderboard is revealed. **Trust your cross-validated AUC.**
- **Watch the filename.** A misnamed file (anything other than `NN_complex_model.pdf` / `NN_complex_model.ipynb` with your group's two-digit number) costs **-10 points** off your M3 grade. Easy points to lose, easier to keep.
