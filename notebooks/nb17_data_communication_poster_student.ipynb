## 3. Context Matters

Every analysis has a goal and an audience. Three rules follow:

- **Separate exploration from final analysis.** Resist the temptation to show everything you did. The reader did not ask for your debugging history.
- **Adapt the report to the audience.** Decision-makers rarely care about execution details. They care about the decision the number supports.
- **Keep one specific learning objective in mind.** It tells you which information is relevant and which is filler.

Isolated numbers say very little. To support evidence-based decisions, an appropriate **basis for comparison** has to be established up front. *"Revenue rose 12%"* means nothing without a peer group, a prior period, or a forecast next to it.

## 4. Visualization Derives From Data

Before you choose a chart, ask three questions about the data behind it:

1. **What type of data is this?** Categorical, ordered, continuous, time-indexed?
2. **How many dimensions are you encoding?** One? Two? Three or more?
3. **Where will the chart be consumed?** Most reports, and every conference poster, are consumed in 2D from three feet away. Encoding more than two dimensions on a static page usually confuses the reader.

And one rule that overrides all three: **be careful with scales.** A bad axis can flip the conclusion of a chart without changing the underlying data.


### 4.1 Use Graphs Instead of Tables

The next two images show identical data: once as a table of numbers, once as a bar chart. Before scrolling down, ask yourself which format lets you extract the comparison in less time. The answer drives a design rule you will use on every poster figure.

<center>
<img src="figures/table_vs_plot.png" width="75%"/>
<br><em>The table technically contains the answer. The plot reveals it at a glance.</em>
</center>


<center>
<img src="figures/table_vs_plot_02.png" width="75%"/>
<br><em>Same numbers, two encodings: the chart wins on every reader.</em>
</center>


When the relationship between numbers is comparative or trend-based, a chart usually beats a table. Reserve tables for cases where readers need to **look up specific values**, not interpret a pattern. *(Source: Kastellec & Leoni, 2007, "Using Graphs Instead of Tables in Political Science.")*

### 4.2 Scales Can Be Misleading

A truncated y-axis is the single most common chart failure in business presentations. The image below starts its y-axis well above zero, turning a modest percentage-point difference into what looks like a dramatic swing. A CFO who sees this chart in a quarterly review will make a different decision than a CFO who sees the same data on a zero-based axis.

<center>
<img src="figures/escala-fail.PNG" width="60%"/>
<br><em>Truncated y-axes turn a 3-point swing into a visual landslide.</em>
</center>


For poster figures, default to **axes that start at zero** for any quantity where zero is a meaningful reference (counts, percentages, currency). When a non-zero baseline is justified (for example, a tightly clustered time series), say so explicitly in the caption.

---


### 4.3 Avoid Dual Axes

Dual-axis charts appear in boardroom slide decks more often than in any other setting, and they mislead in every one. When two series share an x-axis but each gets its own y-axis, the visual alignment between the two lines is controlled by whoever chose the axis ranges, not by the data.

<center>
<img src="figures/eixo-duplo.png" width="55%"/>
<br><em>Two y-axes invite the eye to read causation that the data does not support.</em>
</center>


### 4.4 …And Triple Axes

If dual axes strain the reader's ability to hold two scales in their head, triple axes abandon it entirely. The image below is real, and if you find yourself reaching for a third axis, it is a signal that one chart is trying to tell three stories at once. Split it into three panels.

<center>
<img src="figures/eixo-triplo.png" width="60%"/>
<br><em>If two axes were a stretch, three are a magic trick.</em>
</center>


Dual-axis charts force the reader to mentally align two scales. When the alignment is arbitrary (and it almost always is), the chart hides as much as it reveals. Prefer **two side-by-side panels** with a shared x-axis, or **standardize both series** (e.g., to z-scores or percent-change-from-baseline) and overlay them on a single shared axis.

### 4.5 Fewer Pie Charts

The pie chart is the default choice for "show me a breakdown," and it is almost always the wrong one. The reason is perceptual: the human eye estimates lengths far more accurately than angles or areas. The image below shows the same data as a pie chart and a bar chart side by side: notice how much faster you can rank the categories in the bar version.

<center>
<img src="figures/pie-vs-bar.png" width="82%"/>
<br><em>The pie chart asks the reader to compare angles. The bar chart asks them to compare lengths. Lengths win.</em>
</center>


### 4.6 When 3D Distorts the Comparison

Sometimes the chart type is defensible but the *rendering* sabotages it. The image below is a three-dimensional pie chart Steve Jobs presented at the **Macworld 2008** keynote to show U.S. smartphone market share by brand. The intent was persuasive: convince the audience (technology media who would introduce the iPhone to buyers considering a switch to Apple, and retailers deciding whether to stock it) that the iPhone 3G had become a serious contender, landing in **third place at 19.5%**. The chart does deliver that headline. But three design choices quietly distort the very comparison it asks the reader to make.

<center><img src="figures/steve-jobs-pie.jpg" width="72%"/><br><em>Steve Jobs presents U.S. smartphone market share at Macworld 2008. The 3D tilt inflates the slices nearest the viewer. Source: <a href="https://www.engadget.com/2008-01-15-live-from-macworld-2008-steve-jobs-keynote.html">Engadget, Live from Macworld 2008: Steve Jobs keynote</a>.</em></center>

**What this teaches:** the format is forgivable (a pie *can* show a few brands splitting a whole), but the execution introduces three separate distortions.

First, the **3D effect misleads the eye about rank.** The chart is tilted so the bottom of the pie sits visually closer to the viewer than the top, which inflates the apparent area of the lower slices. The iPhone's green slice ends up looking larger than the slice actually ahead of it, nudging the audience to read the iPhone as *second* place when the numbers put it third. There is no analytical reason to add the third dimension here; its only effect is to flatter the product.

Second, the **slices carry numbers but no category labels.** To learn which product each wedge represents, the reader has to bounce between the wedge and a separate legend, and even then the pie makes the comparison hard. The headline gap is small: third place is 19.5% and second place is 21.2%, a 1.7-point difference that translates into nearly identical angles. A pie can communicate a *large* difference at a glance, but it is the wrong tool for distinguishing two nearly equal shares.

Third, **nothing in the chart encodes the ranking.** The wedges are not ordered ascending or descending, there is no color scale to signal magnitude, and even the legend is left unsorted. The reader is handed the pieces of an ordered comparison without any of the visual cues that would let them see the order.

The fix is structural, not cosmetic: replace the pie with a **horizontal bar chart ordered by share**, with each brand labeled directly at its bar. Lengths along a common axis make a 1.7-point gap legible, and direct labels remove the legend round-trip entirely.

> **A question that often comes up here:** *"is the pie chart really banned?"* Not banned, but reserve it for the rare case of **two or three categories that sum to a meaningful whole**, where the headline is the overall split rather than a fine comparison between slices. For anything else (especially anything ordered or anything with more than three categories) a horizontal bar chart will outperform a pie chart on every reader test we have ever run in this course.

## 5. Less Is More: Separate Signal From Noise

The more visual elements your chart contains, the higher the **cognitive load** on the reader. Every line, label, gridline, and decoration is a small tax. Your job is to lower that tax.

This is what Edward Tufte called the **data-ink ratio**:

$$\text{Data-Ink Ratio} = \frac{\text{Data-Ink}}{\text{Total ink used to print the graphic}}$$

The numerator is the ink that *encodes data*. The denominator is everything you printed. Your goal is to push the ratio toward 1.0 by deleting any ink that does not encode data, and only that ink.

The next nine sub-sections walk a single bar chart through a **seven-step cleanup**, removing one piece of non-data ink at a time. Watch what happens at each step: the underlying numbers never change, but the chart's signal grows louder as the noise around it falls away. By the time we reach the "before and after" panel in §5.9, the same data tells a much clearer story.

### 5.3 Step 2: Eliminate the Border

The border marks the edge of the plot area, but the axis lines and data points already define that boundary. Removing it is the cheapest cleanup: zero information lost, one layer of visual noise gone. In matplotlib: `ax.spines['top'].set_visible(False)` and the same for the right spine.

<center>
<img src="figures/limpeza-2.PNG" width="65%"/>
<br><em>The frame around the plot encodes nothing. Delete.</em>
</center>


### 5.4 Step 3: Clean the Gridlines

Gridlines help when readers need to read exact values off the y-axis. On a poster, the story is almost always about a trend or a comparison, not about precise readoffs. Heavy gridlines compete with the data lines for the reader's attention. Lighten them to a pale grey, reduce their number, or remove them entirely.

<center>
<img src="figures/limpeza-3.PNG" width="65%"/>
<br><em>Heavy gridlines compete with the data. Lighten or remove.</em>
</center>


### 5.5 Step 4: Remove the Points

When a continuous line carries the story, the individual data markers are redundant ink. Every dot restates information the line already conveys. Removing them lets the trend speak without interruption, and frees up visual bandwidth for the annotation you will add in Section 8.

<center>
<img src="figures/limpeza-4.PNG" width="65%"/>
<br><em>When the line is the message, the dots are redundant ink.</em>
</center>


### 5.6 Step 5: Process the Axes

Long tick labels ("January 2024," "February 2024") force the reader to parse dates instead of reading the chart. Shortening or thinning the ticks, and dropping axis lines that carry no information, reduces the cognitive load without sacrificing precision. The rule of thumb: if a tick label is longer than the gap between ticks, it is too long.

<center>
<img src="figures/limpeza-5.PNG" width="65%"/>
<br><em>Strip overlong tick labels, reduce label density, drop axis lines that are not load-bearing.</em>
</center>


### 5.7 Step 6: Adjust the Label

A y-axis label that floats far from the data forces the reader's eye to leave the chart to understand what the numbers mean. Moving the label close to the data it describes (or, better, integrating the unit into the title) cuts that bounce. Every bounce is a moment the reader might lose interest.

<center>
<img src="figures/limpeza-6.PNG" width="65%"/>
<br><em>Move the title and units close to the data they describe.</em>
</center>


### 5.8 Step 7: Adjust Colors

Decorative color draws the eye to the palette, not to the pattern. A single mid-saturation hue says "this is the data line" without competing for attention. Save the bright accent color for later; Section 6 will show you where to spend it for maximum impact.

<center>

**Synthesizing the seven steps.** The cleanup did three things, in order: (i) deleted ink that encoded nothing: borders and gridlines; (ii) replaced redundant ink: points behind bars, axes that repeated the labels; (iii) chose colors and labels that earned their place. Each step is small. Together they double the data-ink ratio without losing a single number.

> **A question that often comes up here:** *“isn’t a stripped-down chart cold or boring?”* It is the opposite. The more non-data ink you remove, the more visually loud the **remaining** ink becomes. The before/after pair above does not lose information; it amplifies the signal that was already there. The chart on the right is what your poster should look like.

## 6. Hierarchy Among Data

Once you have a clean chart, you have to choose **what the reader looks at first**. That is hierarchy. The eye is not democratic; it lands on whatever is loudest. Your job is to make the loudest thing the *right* thing.

On a poster viewed from three feet away, hierarchy is the difference between a judge who spots your headline number in five seconds and a judge who gives up after ten. The next exercise demonstrates the principle with a task you can feel.

<center>

Hierarchy works because the reader's eye **filters before it counts**. The same idea drives every good poster figure: the headline number should be visually **louder** than every other element, by a factor large enough that you can spot it from across a room. If you remember nothing else from this section, remember this: the reader will look at the loudest thing on your chart whether you designed it that way or not. You are not choosing whether hierarchy exists; you are choosing whether you control it.

### 6.6 Use Colors Strategically

Most categorical color palettes include one saturated accent and several muted tones. Spend the accent on the thing you want the reader to remember. If you have spent it on three different elements, none of them is the accent anymore.

> **A question that often comes up here:** *"how many colors should a poster figure have?"* For headline figures: **two**. One muted base color (greys, light blues) for everything that is context, and **one** accent color for the number you want the reader to walk away with. If a chart has more than three colors, ask whether each one is doing structural work, and if it isn't, pick the first three and grey out the rest.

## 7. Beauty Counts: Not All Data Are Equally Important

Hierarchy chooses what is loudest. Beauty makes the chart pleasant enough that the reader **stays** long enough to hear the message. The two work together: a hierarchical chart that is ugly still loses readers; they flinch away before the data registers. A beautiful chart with no hierarchy buries its point under polish; the reader enjoys the look but walks away without a number.

Beauty in data visualization is not decoration. It is the set of choices (alignment, whitespace, typeface, color temperature) that signal competence and reduce cognitive friction. A well-designed chart tells the reader "this team took the work seriously" before a single number is read. On a research-poster floor where judges evaluate ninety submissions in an afternoon, that signal earns an extra fifteen seconds of attention, which is often the difference between a glance and a conversation.

Pick one element (the headline bar, the title banner, or the key series) and let it carry the brand. Everything else stays neutral.

## 9. Why a Poster Presentation?

The principles above apply to any analytics deliverable: a Jupyter notebook, a Slack message, an executive slide. From here on, we narrow the focus to the specific Final Project deliverable: a single poster, designed for an Undergraduate Research Conference.


<center>
<img src="figures/project-history.jpg" width="55%"/>
</center>

A research poster does five things at once:

- **Showcases your group’s work** in a single static frame any reader can scan
- **Organizes your ideas and results** into the eleven-section architecture used by every academic field
- **Is visually appealing** enough that strangers stop walking past it
- **Encourages interactive discussion**: the poster is a conversation starter, not the conversation itself
- **Demonstrates your mastery** of predictive-analytics concepts in a form non-specialists can evaluate

**Why this list is the spec, not just a description.** The first three items are about the *deliverable*: what the poster has to be. The last two are about the *audience*: what the poster has to do for a reader who cannot ask questions. A conference-floor reader cannot scroll back, cannot ask you to clarify, and will not stay if the opening thirty seconds fail. Every later section in this notebook addresses one of those five constraints.


> **A question that often comes up here:** *"why a poster instead of a slide deck or a written report?"* Posters are the conference-research format precisely **because** they force the constraints we have been training against in this notebook: single page, three-foot reading distance, ninety-second attention budget, no presenter required (the poster has to defend itself even when nobody is standing next to it). Slide decks let you hide behind narration. Written reports let you hide behind length. A poster forces you to commit.

## 10. Designing a Predictive Analytics Poster: Objectives

Before you place a single text block, lock the **three things** the poster has to communicate. These three items are not optional, and they are not interchangeable. They follow a sequence that mirrors the way a non-expert reads evidence, and violating that sequence is the fastest way to lose your reader.

<center>
<img src="figures/paper-abstract.jpg" width="55%"/>
</center>

1. **Communicate the key findings and impact** of your project: the one number a stranger should remember.
2. **Highlight the predictive approach**, the methodology, and any novel insight.
3. **Highlight the business implications**: translate your model output into the decision a stakeholder can now make differently.

If any of those three is unclear to your group right now, the poster is not yet ready to be designed. Stop and align in writing first; the next sections assume the answers exist.

**Notice the sequence: finding → method → implication.** A poster that opens with the model architecture before naming the problem loses the reader who does not yet know why they should care. A poster that lists the business implication before showing the evidence loses credibility; the reader has nothing to anchor the claim to. The order is not negotiable, even on a single 36×48 inch panel; the eye reads top-to-bottom and judges read in that order.

## 12. Planning Your Poster: Visual Hierarchy and Layout

A poster has hierarchy at **three layers**: the layout (what is where), the typography (what is loudest), and the content (what is most important). All three have to agree, and they fail independently.

Think of it this way: a beautifully typeset poster with a single-column layout buries the headline below the fold. A three-column poster with inconsistent fonts makes the reader re-calibrate their eye on every section. And a clean layout with consistent fonts but unfocused content gives the reader nothing to remember. Your job is to get all three layers pulling in the same direction.

<center>
<img src="figures/poster_template.png" width="68%"/>
<br><em>Course poster template: three-column layout with clear visual hierarchy.</em>
</center>

- Provide a visually engaging, **easy-to-navigate** summary.
- **Always lay out in columns**, typically three for a 36×48 inch poster.
- **Headings:** Large and bold, in a consistent hierarchy that guides the reader.
- **Color and contrast:** Pick a simple palette (one accent + neutrals) that highlights the main points without competing with your figures.
- **Font size:** Body text should be legible from **three feet away**; that is the test, not “looks fine on my laptop screen”.
- **Flow:** Logical reading order, top-left to bottom-right.

**The three layers fail independently.** Great content with a one-column layout buries the headline. A three-column layout with inconsistent fonts forces the reader's eye to keep recalibrating. Clean layout and consistent typography but unfocused content gives the reader nothing to remember. Audit each layer separately: print a thumbnail at 25% scale and check whether the headline, the method, and the recommendation are visible in under 10 seconds. If not, one of the three layers is failing, usually typography first, content second, layout third.


### Layout and Design Guidelines

The guidelines below are the mechanical decisions (spacing, font sizes, color counts) that implement the three-layer hierarchy above. Treat them as defaults you can adjust, not rules you cannot break. But if you break one, know which layer of hierarchy you are affecting.

**Keep it simple**

- Limit text and rely on **figures, charts, and bullet points**.
- Use consistent **color schemes**, fonts, and alignment.
- Use a **QR code** to drive your audience to a dashboard, project webpage, or live demo.

**Use of space**

- **White space is your friend**: avoid clutter.
- Group related sections in boxes or visually distinct areas.
- Aim for roughly **60% graphics / 40% text**.

**Key text guidelines**

- **Title:** \~85–100 pt
- **Headings / Subheadings:** \~46–55 pt
- **Body Text:** \~34–38 pt
- **Captions:** \~26–28 pt
- All of the above must be readable from three feet away.

## 14. Designing for Predictive Analytics: The Special Case

Most poster advice is the same for every field, whether the project sits in biology, engineering, or economics. This section is about what changes *because* your project is a modeling project. These are the details a judge who knows predictive analytics looks for first, and they fall into two groups: how you **report your model's performance**, and how you **show that your evaluation was honest**.

**Reporting performance: three things judges look for.**

First, **name your model in plain text.** Write "logistic regression," "random forest," or "gradient boosting" somewhere obvious. Do not make the reader reverse-engineer the method from the metrics.

Second, **report the headline metric with its confidence interval, not a lone number.** A confidence interval is the range your score would likely land in if you ran the study again, and you already computed it with cross-validation back in notebook 08. A judge who sees "ROC-AUC = 0.87" with nothing next to it reads it as overclaiming. "ROC-AUC = 0.87 (95% CI 0.84 to 0.90)" reads as careful work.

Third, **show the top three to five features** that drove the predictions. A horizontal bar chart of feature importance is the right shape, because the reader can scan the ranking from top to bottom without tilting their head.

**Proving the work is honest: two more things.**

Fourth, **label every axis, and put labels directly on the chart** instead of in a separate legend when you have four or fewer lines or bars. The reader is three feet away, not three inches, so the hierarchy and storytelling ideas from Sections 6 and 8 matter even more here.

Fifth, **show your train/test split as a picture.** A small pipeline diagram, or even a simple flow arrow, makes clear that your evaluation never peeked at the test data. Judges trained in analytics look for this the way an auditor looks for the independence note in a financial statement: it is the signal that the numbers can be trusted.

**The two mistakes that hurt analytics posters most.** The first is reporting a single number with no confidence interval, which reads as overclaiming. The second is hiding the baseline, which lets the judge assume your model barely beats a coin flip, because you never showed what you were beating. Both are quick fixes: the confidence interval came out of notebook 08, the baseline came out of notebook 09, and the rubric gives you visible credit for showing them.

> **A question that often comes up here:** *"why do judges care so much about the confidence interval and the baseline?"* Because together they answer the only two questions a skeptical reader really has: "is this result real, or did you just get lucky on one split?" and "is it actually better than the obvious, do-nothing option?" The confidence interval answers the first by showing how much your score wobbles from fold to fold; the baseline answers the second by giving the judge something to compare against. A strong score with no interval and no baseline looks like a claim with no evidence behind it, no matter how good the number is.

The poster is the medium. The story is the thing. Two reminders before you start writing: one about *tone*, one about *discipline*.

A poster amplifies the cost of breaking them: every adjective competes for the reader's 90-second budget, and every nominalization makes a sentence one read-through longer.

## 16. Effective Figures and Tables on a Poster

Everything you learned in Sections 3 through 8 applies here, with one catch: a poster is read from three feet away, on the first try, with no second chances. Your reader cannot squint at a tiny axis label. They cannot hover over a point for a tooltip. They cannot scroll back to re-read a caption. So every figure has to land at arm's length, the first time someone looks at it.

For **charts and graphs**, three rules do most of the work.

First, **label every axis.** An unlabeled axis is a question the reader has to answer before they can even begin reading your chart.

Second, **use one accent color** for the data that makes your point, and let everything else fade to grey. The reader's eye should travel straight to the story.

Third, **write captions that state the conclusion, not the topic.** "XGBoost beats the baseline by 8 points of ROC-AUC" tells the reader what to think. "ROC-AUC comparison across models" just repeats the axis label and wastes the space.

For **tables**, there is one test: could a chart make the same point faster? If yes, use the chart. A table earns its place only when the reader needs to look up an exact value. A small comparison table (four rows, three columns) can work, and so can a confusion matrix with one cell highlighted. But a table with eleven rows and eight columns belongs in your analysis notebook, not on a poster a judge reads in ninety seconds.

**Make it big enough to read across a room.** On a poster, figure labels should be at least 24 pt and captions at least 18 pt. Here is the test: print one figure at full size, tape it to the wall, and step back three feet. If you cannot read the axis label from there, neither can the judge.

> **A question that often comes up here:** *"how many figures should the poster have?"* Three to five is the sweet spot. Aim for one headline figure that captures the main result (your cross-validation comparison chart, or the confusion matrix with the key cell highlighted), one method figure (the pipeline diagram, or a simple picture of how you split the data), and one diagnostic figure (a calibration plot, a feature-importance bar chart, or an error breakdown by customer segment). Past five figures, they start fighting each other for attention, which is the same hierarchy problem from Section 6, now scaled up to the whole poster.

**Extend the poster beyond the wall with a QR code.** A poster holds about 600 words and a handful of figures, but your project is bigger than that. A QR code is the small square barcode anyone can scan with a phone camera to open a web link in seconds. Print one in a corner of the poster and you hand curious readers a door to everything that did not fit: an interactive dashboard, a live demo app, your project's code repository, a short video walkthrough, or the full analysis notebook. You can even point it at your team's LinkedIn pages, so a recruiter who liked your work can connect with you before they walk away.

A few practical tips make QR codes work. Add a one-line label under the code so the reader knows what they are scanning ("Scan for the live dashboard" beats a bare square). Use any free QR generator, keep the link short and stable, and test the printed code with two or three different phones before you submit. And treat the code as a bonus, not a crutch: the poster still has to tell the whole story on its own, because most readers will never scan it.

> **A question that often comes up here:** *"what should my QR code point to?"* Pick the one destination that rewards a curious reader the most. If your team built an interactive dashboard or an app, link there, because it lets the reader play with your model instead of only reading about it. If you did not build a live tool, link to your code repository, a two-minute video, or your team's LinkedIn pages for networking. Just avoid dropping people onto a raw, unlabeled file: whatever loads on their phone should look as polished as the poster that sent them there.

## 17. Results and Interpretation

This is the section judges study the longest, so make it easy on them. Show three things, in this order.

First, the **headline number**: the single metric that says whether your model is actually useful.

Second, the **comparison** that proves the number is real and not luck. This is almost always your model shown next to a simple baseline, with both confidence intervals so the reader can see the gap is solid.

Third, a **diagnostic** that shows you took the result seriously: a calibration plot, a fairness check across groups, an error breakdown by segment, or a confusion matrix with one cell highlighted.

That is the whole Results block: three figures, one short paragraph each. The temptation is to show every metric from every notebook, but that buries the reader. Pick the three figures that tell the clearest story, and let the rest stay in your analysis notebook.

**Showing the performance.** The clearest pattern is a side-by-side or grouped-bar comparison. For a classification project, use a confusion matrix and highlight the cell your stakeholder cares about most: usually the false negatives when missing a real case is expensive (medical screening, fraud detection), or the false positives when a false alarm is expensive (spam filters, loan approvals). For a regression project, use an actual-versus-predicted scatter plot, or a bar chart of errors broken out by group. In every case, place your model right next to the baseline from notebook 09 (the dummy classifier, or the mean-only predictor) on the same figure, so the judge sees the improvement at a glance.

**Showing your thinking.** Judges reward this more than most teams expect. They want to see that you can explain *why* the model behaved the way it did, not just report a score. Which features mattered most? Where does the model struggle, and for which customers or cases? What would you fix with more data, more time, or a different method? Naming your model's weak spot is not a confession; it is a credibility builder. The team that hides its worst-performing segment is exactly the team a sharp judge catches.

**The pattern in one line: show the confidence interval, name the baseline, label the failure mode.** The confusion matrix with one cell highlighted tells the cost story. The comparison bar chart against the baseline tells the "is this real?" story. The calibration plot, fairness check, or segment breakdown tells the "where does it fail?" story. Three figures, a few sentences each, and the honesty is what makes the whole thing believable.

> **A question that often comes up here:** *"should I put the confusion matrix or the ROC curve on the poster?"* For a classification project, lead with the confusion matrix and highlight one cell (false negatives or false positives, depending on which mistake costs your stakeholder more). The confusion matrix tells a cost story anyone can read, while the ROC curve tells a threshold story that takes training to interpret, so it rarely earns its space on a poster. For a regression project, show the actual-versus-predicted scatter or the residual plot, whichever makes the "where does the model miss?" story clearest.

## 18. Conclusion and Future Work

The two short paragraphs at the bottom of the poster carry more weight than their size suggests. The **Conclusion** is the line a reader repeats to a colleague the next day, so it has to be concrete and memorable. **Future Work** is the line that shows your team can see past the assignment to the next step. Both are short, roughly 50 words each, and both deserve more editing than anything else on the poster.

**A strong Conclusion does three things in three sentences.** It restates the main finding in plain words a non-expert can repeat ("the model flags the riskiest 10% of accounts with 87% recall"). It names the real-world payoff ("that gives the outreach team a three-week head start before renewals close"). And it says how sure you are ("the 95% cross-validation interval of 0.84 to 0.90 says this is not a lucky split"). Set that against a weak conclusion like "the model performs well," which is true, forgettable, and useless to the stakeholder.

**A strong Future Work section names specific next steps, not wishes.** "Adding lag features to capture seasonal demand (see notebook 16) would help the model anticipate spikes" is concrete and credible. "A more robust model could be built" is filler that signals you ran out of ideas. The best future-work items come straight from limitations you already found: a weak segment from your error analysis, a data-quality gap from your EDA, or a population shift your monitoring plan would catch.

**The test for both paragraphs: specific beats vague, every time.** "The model performs well" is forgettable. "The calibrated model supports a screening cutoff of 0.42, catching 85% of true cases at a 5-to-1 cost ratio" is a line a reader can carry into their next meeting. Spend more time on these two paragraphs than on any equation on the poster, because they are what a busy reader actually remembers.

> **A question that often comes up here:** *"how do I write the conclusion if my final numbers might still change?"* Draft it now anyway. The shape of the sentence (finding, then payoff, then confidence level) does not change even if a metric shifts by a point or two once your champion model is locked. Swapping in the final number takes thirty seconds; writing a clear paragraph from scratch on submission night takes an hour you will not have.

## 19. Final Touches

Acknowledgments, references, and proofreading are not afterthoughts; they are free rubric points. Judges check your reference list to confirm you credited the tools you used (`scikit-learn`, `pandas`, `statsmodels`), your data source, and the course materials behind your method. A missing citation or a typo in the title is a small mistake, and a small mistake is the easiest kind to prevent.

**Acknowledgments and references are scored.** List the libraries you relied on (`scikit-learn`, `pandas`, `statsmodels`), your data source, and the textbook or course materials that shaped your method. Pick one citation format and stick to it (APA is fine for this course). Thank the faculty, teammates, and any funding source that helped along the way.

**Proofreading and rehearsal are where the last few points hide.** Spell-check every block of text. Check every number against your notebook: a poster that says "ROC-AUC = 0.87" when your notebook says 0.85 quietly undermines the whole project. Make sure every image is sharp at print size, because a blurry confusion matrix blown up to poster scale looks worse than no figure at all. Then practice out loud on a non-expert, such as a roommate, a parent, or anyone outside the course. If they cannot repeat your main conclusion in one sentence, the poster is not finished yet.

**The rehearsal is the highest-value step in this whole list.** Print a draft poster at full size, tape it to a wall, stand three feet back, and time how long it takes a teammate who did not build the project to find your headline number. If they cannot find it in 30 seconds, the layout has buried it, and no amount of proofreading will fix that. Do this at least 48 hours before the deadline, so you have time to redesign rather than just retype.

## 20. Presenting Your Poster (Optional for the Summer Course)

If your group elects to present at the Undergraduate Research Conference, the poster is half the work; the live presentation is the other half. The two demand different skills: the poster is a self-standing document that works when nobody is there; the presentation is a guided tour with a human voice. Tips below come from past course teams who have presented at URC; the advice is concrete because it comes from real experience, not from a textbook.

The mechanics of a good poster presentation reduce to a few high-leverage habits:

- **Prepare three pitches ahead**: **30-second, 90-second, 2-minute**. Use the poster as the visual guide for each. The 30-second version is a single sentence: "We built a model that predicts X with Y% accuracy, and the business implication is Z." The 90-second version adds the method and the key limitation. The 2-minute version walks the poster top to bottom.
- **Prepare questions**: open with *"Are you familiar with...?"*. Actively ask questions of your audience to keep it a conversation.
- **Arrive early** to set up; board assignments shift sometimes.
- **Stay close** to your poster, off to the side. Do not block the figures.
- **Use your hands** to direct the listener's eye to specific sections of the poster.
- Do not let one teammate dominate; rotate naturally.
- [**Follow the dress code: business professional**](https://www.purdue.edu/undergrad-research/conferences/fall/photo-gallery.php). [CCO Career Closet](https://www.cco.purdue.edu/Students/WhatWeOffer?tab=CareerCloset) can help if you need it.

## 22. Wrap-Up: Key Takeaways

**The Six Principles you can audit any chart against:**

1. **Context Matters**: every number needs a baseline.
2. **Visualization Derives From Data**: the data shape picks the chart, not the other way around.
3. **Less Is More**: high data-ink ratio beats decorative ink, every time.
4. **Hierarchy**: the eye is not democratic; choose what wins it.
5. **Beauty Counts**: pleasant charts retain readers long enough to deliver the message.
6. **Telling Your Story**: every figure should pass the cover-and-restate test.

**The Eleven Poster Sections you can plan against:**

Title → Abstract → Business Problem → Analytics Framing → Research Question → Data/EDA → Methodology → Model Building → Business Implication → Conclusion → Acknowledgments.

**One sentence to take into the final-poster sprint:**

> *“Cover up your favorite figure. If a stranger cannot restate the conclusion in one sentence, the figure is not yet ready.”*
