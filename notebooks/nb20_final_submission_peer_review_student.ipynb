### 1.3 The Poster: Required Components

A standout poster reads in roughly five minutes. Yours must carry, in a logical Intro → Methods → Results → Conclusions flow:

1. **Project title:** concise and informative. *If your project uses a synthetic generated dataset, the title must say so.*
2. **Group members + section.**
3. **Prediction problem:** framed as a question with a question mark (e.g., *"Can six-month customer-churn risk be predicted from transaction history?"*).
4. **Motivation and significance:** why this prediction matters.
5. **Data overview:** source, size, key variables, and the response-variable type.
6. **Methodology:** the leakage-safe preprocessing pipeline (`ColumnTransformer` + `StandardScaler` + `OneHotEncoder`), feature engineering, the M2 baseline and the M3 complex model, the cross-validation protocol, and the hyperparameter search.
7. **Results, including the one-shot test-set evaluation (required; see §1.5):** the M3 champion's CV mean ± 95% CI, the test-set metric opened exactly once, and the INSIDE / ABOVE / BELOW verdict.
8. **Interpretation and insights:** feature importance or coefficients, and what they say about the business question.
9. **Limitations + next steps:** honest about what the data can and cannot support.
10. **References:** datasets, papers, external resources. (The submitted `NN_final.ipynb` is your canonical reproducibility artifact; no separate code-repository link is required on the poster.)

### 1.4 Required Visualizations on the Poster

A great poster carries **4–6 high-information figures, not 12 mediocre ones**. Every figure needs axis labels, units where they apply, a clear legend, and a one-sentence caption that names the takeaway. The following figures are **required** (omitting any of them costs rubric points under Visual Design and Results):

- **Model-comparison bar chart:** the M2 baseline vs. the M3 champion (and any further-tuned variant), each as a bar with **95% Student's *t* CI error bars**. This is the figure that visually justifies your CI-overlap-rule decision.
- **Test-set verdict figure:** the test metric overlaid on the M3 CV CI bar, picturing the INSIDE / ABOVE / BELOW verdict from §1.5.
- **Feature-importance / coefficient plot:** a horizontal bar chart of the champion's top features (signed coefficients for linear/logistic models; permutation or impurity importance for tree-based models).
- **For regression projects:** a predicted-vs-actual scatter with the 45° reference line, OR a residual plot with a `y = 0` reference line.
- **For classification projects:** a confusion matrix at your operating threshold AND a ROC curve with AUC, OR a Precision–Recall curve with PR-AUC (lead with PR if the positive class is rare).

If a figure tells the same story as one already on the poster, replace it. Every figure should add information.

### 1.5 The One-Shot Test-Set Evaluation (the lockbox ceremony)

This is the single most important, and most easily botched, step of Milestone 04. Since nb01 your group has kept its test set in a **lockbox**: M1 framed the question, M2 and M3 selected models entirely on **cross-validation**, and the test set was never touched. M4 is the **one and only** authorized opening of that box.

The ceremony has exactly four moves:

1. **Reload the champion you saved at M3.** Use the `champion_pipeline.joblib` from your M3 final-training step (retrained on the full training fold). Do **not** re-tune anything here.
2. **Open the locked test set once.** Compute the champion's metric on the held-out test set, using RMSE/MAE/R² for regression and ROC-AUC/F1 for classification. A single evaluation.
3. **Render the verdict against the M3 CV 95% confidence interval:**
   - **INSIDE:** the test number falls within the M3 cross-validation CI. The model generalized exactly as the CV predicted. This is the outcome you want.
   - **BELOW:** the test number is *better* than the CV expected (lower error, or higher score). Pleasant, but report it plainly rather than celebrating.
   - **ABOVE:** the test number is *worse* than the CV upper bound. The model underperformed its CV estimate; add a one-sentence diagnosis and a clear **deploy / pause / re-train** recommendation.
4. **Report it on the poster and in `NN_final.ipynb`:** the test metric, the CV CI, and the verdict, side by side (this is the test-set verdict figure from §1.4).

> **A question that often comes up here:** *"Our test number came out ABOVE the CV interval. Can we just tune a bit more and re-open the test set?"* No. The moment you select anything against the test set, it stops being a held-out exam and the lockbox discipline that carried through M1 → M2 → M3 is gone, and graders can usually see it in the notebook's cell history. An ABOVE verdict is not a failed project; it is an **honest, reportable finding**. State the gap, give a one-sentence diagnosis (overfitting, distribution shift, an unlucky small test split), and make a deploy/pause/re-train call. A team that reports an honest miss with a recommendation scores better than a team that quietly re-opened the box to make the number look good.

## 2. The Course Case Competition: Final Kaggle Submission

The **Course Case Competition (Kaggle Bank Churn)** runs in parallel with the Final Project and uses the **same CI-first workflow** you built for M3. At course close your group locks in its final Kaggle submission and turns in one notebook.

**The task.** For each customer in the test set, predict the **probability of churn** (`Exited = 1`). Submissions are scored by **ROC-AUC**. The **public** leaderboard scores a fraction of the test set during the competition; the **private** leaderboard, revealed at the close, is what grades you.

**What to submit.**

- **On Kaggle:** your team's final submission `.csv`: two columns (`id`, `Exited`), every test row covered, probabilities in [0, 1].
- **On Brightspace:** **`NN_kaggle_code.ipynb`**: the fully replicable Colab notebook for your best model (loading → preprocessing → feature engineering → tuning → final fit → writing the submission file). One member uploads for the team. A filename that doesn't follow the `NN_kaggle_code.ipynb` convention costs **-10 points**.

**The rules that quietly cost teams grades:**

- **Name your Kaggle team exactly `Group NN`** (with the leading zero, e.g., `Group 03`) so standings map back to the gradebook, and form the team **before anyone submits**. A leaderboard team name that does not match `Group NN` at the deadline costs **-10 points**.
- A team of four **shares 5 submissions per day**, not 5 each. Spend them on experiments that actually teach you something.
- **Do not chase the public leaderboard.** Pick your final submission on the most defensible **cross-validated AUC**, not the highest public score. Models that overfit the public split routinely drop 10+ ranks when the private leaderboard appears.

**How it's graded.** Your competition grade is **60% leaderboard performance + 20% team participation + 20% peer evaluation, minus penalties.** The **60 leaderboard points** come from your rank on the **private** leaderboard at the close: the top team earns the full 60, every rank below steps down by the same modest amount, and a model no better than random guessing (AUC 0.5) earns zero — so rank, not a hair's-width AUC gap, moves your grade. **Team Participation (20 points)** evaluates your submitted, reproducible notebook: a Run-All notebook (4), a leakage-safe pipeline (4), cross-validated evaluation of a baseline and a complex model with 95% CIs (4), systematic tuning with the final model chosen by the CI-overlap rule rather than the public leaderboard (4), the required hyperparameter-search and CV-CI comparison figures (2), and a short write-up with your AI-use note (2); you earn it as long as your team makes at least one valid Kaggle submission and turns in the notebook, regardless of where you rank. **Peer Evaluation (20 points)** is the one individual piece — your teammates' confidential ratings set it through the separate Peer Evaluation assignment, so pulling your fair share keeps the full 20%. Two **-10 penalties** apply on top, both noted above: a Kaggle team name that is not `Group NN`, and a notebook filename that is not `NN_kaggle_code.ipynb`.

> **A question that often comes up here:** *"Why might our leaderboard rank change when the competition closes?"* Because the two leaderboards score **different slices of the same test file**. While the competition is open, the **public leaderboard** scores only about **20%** of the test rows, and that small slice is all you can see. When the competition closes, the **private leaderboard** is revealed and scores the remaining **\~80%**, and *that* is the number used for grading. A model you tuned to climb the public board has effectively been fit to that 20% slice, so it can sit high during the competition and then **drop several ranks** once the unseen 80% is scored. Groups routinely shuffle positions at the close for exactly this reason. That is why you ship the submission with the most defensible **cross-validated** AUC, not the highest public score: your CV confidence interval estimates how the model performs on data it has never seen, which is precisely what the private 80% is. Treat the public board as a sanity check that nothing is badly broken, never as the thing you optimize. The worked starter pipeline from **nb18** walks the full pattern end to end on this dataset.

## 3. Course Closeout: Peer Evaluation and Reflection Survey

Two individual closeouts mark the course complete, submitted by **each of you individually** (not one per group), in the last Brightspace module. The first is the Peer Evaluation; the second is the reflection survey.

### 3.1 Peer Evaluation Submission

You also need to complete the **Peer Evaluation Submission: Final Project and/or Course Case Competition**, the confidential form you'll find in its own course assignment on the Brightspace page. It is its own **individual** grade component for *both* the Final Project and the Course Case Competition, and it can **significantly affect your individual grade**, so it is worth a few minutes of honest attention.

**How you submit it.** You fill the form out **once for each teammate** you evaluate (you do not evaluate yourself), and for each one you choose whether it counts toward the Final Project, the Course Case Competition, or both. Rate honestly: uniform high scores for everyone carry no penalty and no benefit. The math only moves when you actually distinguish one teammate from another.

**How it sets your grade.** Your individual peer-evaluation grade is **adjusted by the average share of the contribution your teammates reported for you**. Pull your fair share and you keep the full component; if the teammates who evaluated you report that your contribution fell clearly below the group, your individual points for that component come down.

**What skipping it means.** Submitting the form is the only way *your* read of who actually carried the work gets counted. If you do **not** submit it, you are effectively reporting that **every one of your teammates deserves the full points** on the peer-evaluation rubric, for **both** the Course Case Competition and the Final Project, because a teammate you do not evaluate receives full credit from you by default. So if the contributions on your team were not equal, the only way to say so is to submit the form.

**Bottom line:** submit the Peer Evaluation for both components, rate honestly, and remember that your own individual grade reflects the contribution share your teammates report for you.
